```python
import pandas as pd
import numpy as np

# --- 1. TES POSITIONS ACTUELLES (Exemple) ---
# Imagine que tu as ces montants en $ dans ton compte aujourd'hui
portefeuille_actuel = {
    "SPY": 2500,
    "GLD": 1800,
    "TLT": 2200,
    "AAPL": 3100,
    "XLE": 400
}

# --- 2. L'ALLOCATION CIBLE (Issue de ton optimisation précédente) ---
allocation_cible = {
    "SPY": 0.22,
    "GLD": 0.18,
    "TLT": 0.25,
    "AAPL": 0.15,
    "XLE": 0.20
}

def generate_rebalancing_orders(current_p, target_a):
    df = pd.DataFrame([current_p]).T
    df.columns = ['Valeur_Actuelle']
    
    total_value = df['Valeur_Actuelle'].sum()
    df['Poids_Actuel'] = df['Valeur_Actuelle'] / total_value
    df['Poids_Cible'] = pd.Series(target_a)
    
    # Calcul de l'écart
    df['Ecart_Relatif'] = df['Poids_Cible'] - df['Poids_Actuel']
    df['Montant_Ajustement'] = df['Ecart_Relatif'] * total_value
    
    # Identification des ordres
    df['Action'] = df['Montant_Ajustement'].apply(lambda x: 'ACHETER' if x > 0 else 'VENDRE')
    
    print(f"💰 VALEUR TOTALE DU PORTEFEUILLE : {total_value:.2f}$")
    print("-" * 50)
    print(df[['Poids_Actuel', 'Poids_Cible', 'Action', 'Montant_Ajustement']])
    print("-" * 50)
    
    return df

# Lancement
ordres = generate_rebalancing_orders(portefeuille_actuel, allocation_cible)
```
