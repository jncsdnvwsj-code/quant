```python
import yfinance as yf
import pandas as pd
import numpy as np

tickers = ["AAPL", "TSLA", "MSFT"]
df = yf.download(tickers, period="3y", auto_adjust=False)
prices = df['Adj Close'].tail(730)

log_returns = np.log(prices/prices.shift(1)).dropna()

mean_return_vector = log_returns.mean()
print(mean_return_vector)

covariance_matrix = log_returns.cov() * 252
L = np.linalg.cholesky(covariance_matrix)

Z = np.random.normal(0, 1, size=(3, 1))
X = np.dot(L, Z)
cov_X = 0.5 * np.dot((X - np.mean(X, axis=0)).T, X - np.mean(X, axis=0))
```
