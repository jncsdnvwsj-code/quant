```python
import yfinance as yf
import pandas as pd

# --- CONFIGURATION DE TON PORTEFEUILLE ---
# Remplace 'Prix_Achat' par le prix réel auquel tu as acheté tes parts
MY_PORTFOLIO = {
    "NVDA": {"parts": 28, "achat": 135.00},
    "NRG":  {"parts": 21, "achat": 105.00},
    "ISRG": {"parts": 3,  "achat": 490.00},
    "GS":   {"parts": 2,  "achat": 520.00},
    "MSFT": {"parts": 3,  "achat": 415.00},
    "CMC":  {"parts": 17, "achat": 65.00}
}

STOP_LOSS_PCT = 0.15  # 15% de sécurité

def run_sentinel_dashboard():
    tickers = list(MY_PORTFOLIO.keys())
    data = yf.download(tickers, period="1d")['Close'].iloc[-1]
    
    total_investi = 0
    total_actuel = 0
    
    print(f"\n{'TICKER':<8} | {'PRIX ACTUEL':<12} | {'GAIN/PERTE':<12} | {'STATUT'}")
    print("-" * 55)
    
    for t in tickers:
        p_achat = MY_PORTFOLIO[t]['achat']
        p_actuel = data[t]
        parts = MY_PORTFOLIO[t]['parts']
        
        perf = (p_actuel / p_achat) - 1
        total_investi += p_achat * parts
        total_actuel += p_actuel * parts
        
        # Statut de sécurité
        status = "✅ OK"
        if perf <= -STOP_LOSS_PCT:
            status = "⚠️ STOP-LOSS ATTEINT"
        elif perf > 0.10:
            status = "🚀 PERFORMANCE ++"
            
        print(f"{t:<8} | ${p_actuel:>10.2f} | {perf:>10.2%}")

    perf_globale = (total_actuel / total_investi) - 1
    print("-" * 55)
    print(f"💰 VALEUR TOTALE : {total_actuel:.2f}€ ({perf_globale:+.2%})")
    print(f"🛡️ PROTECTION    : Stop-Loss global activé à 15%")

if __name__ == "__main__":
    run_sentinel_dashboard()
```
